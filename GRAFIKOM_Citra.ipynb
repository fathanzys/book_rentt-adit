```python
import cv2 as cv
import numpy as np

from matplotlib import pyplot as plt
homies = cv.cvtColor (cv.imread('homies.jpg'), cv.COLOR_BGR2RGB)
homies = homies[:1600,:1600]

abu_homies = cv.cvtColor(cv.imread('homies.jpg'), cv.COLOR_BGR2GRAY)
abu_homies = abu_homies[:1600,:1600]
```

```python
print("Banyaknya piksel pada nol pada gambar ini: {0}".format(np.mean(homies==0)))
plt.figure(figsize = (20,12))
plt.imshow(homies)
plt.show()

print("resolusi gambar", homies.shape)
print("tipe data: ", homies.dtype)
```

```python
print("Banyaknya piksel pada nol pada gambar ini: {0}".format(np.mean(abu_homies==0)))
plt.figure(figsize = (20,12))
plt.imshow(abu_homies)
plt.show()
plt.gray()

print("resolusi gambar", abu_homies.shape)
print("tipe data: ", abu_homies.dtype)
```
